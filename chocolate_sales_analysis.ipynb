## 🧹 Data Cleaning

The following cleaning steps are performed:
- Removed duplicate records
- Handled missing values
- Corrected data types
- Renamed columns where necessary

```python
df.columns
```

```python
df.columns = df.columns.str.lower()
df.rename(columns={'amount': 'revenue'}, inplace=True)
```

```python
df.columns
```

```python
df['revenue'] = df['revenue'].str.replace('$', '', regex=False)
```

```python
df['revenue'] = df['revenue'].astype('float')
```

## Missing Value Treatment

### Missing Values Identified

| Column | Missing Values | Treatment |
|----------|----------|----------|
| Order_Date | 437 | Rows removed using `dropna()` |
| Discount_Pct | 489 | Missing values filled with `0` |
| Price_per_Box | 457 | Calculated using `Amount / Boxes_Shipped` |
| Marketing_Spend | 461 | Missing values filled with `0` |

### Reasoning

- `Order_Date` contained 437 missing values out of 200,000 records (0.22% of the dataset). Since the percentage was very small, the affected rows were removed.
- Missing values in `Discount_Pct` were replaced with `0`, assuming no discount was applied.
- Missing values in `Marketing_Spend` were replaced with `0`, assuming no marketing spend was recorded.
- Missing values in `Price_per_Box` were derived using the business rule:

  `Price_per_Box = Amount / Boxes_Shipped`

  This allowed missing values to be reconstructed rather than using an average or median value.

```python
df['price_per_box'] = df['price_per_box'].fillna(
    df['revenue'] / df['boxes_shipped']
)
```

```python
df['discount_pct'] = df['discount_pct'].fillna(0)
```

```python
df['marketing_spend'] = df['marketing_spend'].fillna(0)
```

```python
df = df.dropna(subset=['order_date'])
```

```python
df.isnull().sum()
```

```python
df['order_date'] = pd.to_datetime(
    df['order_date'],
    format='mixed',
    dayfirst=True,
    errors='coerce'
)
```

```python
df['order_date'].dtype
```

```python
df['salesperson'] = df['salesperson'].str.strip()
```

```python
df['country'] = df['country'].str.title()
```

```python
df.columns
```

## Date-Based Columns

To support time-series analysis and dashboard development, additional date-related columns were derived from the `order_date` field after converting it to the `datetime64[ns]` data type.

### New Columns Created

| Column Name | Description |
|------------|-------------|
| `year` | Extracted year from the order date |
| `month_num` | Month number (1–12) used for chronological sorting |
| `month_name` | Month name used for reporting and visualizations |
| `weekday_num` | Weekday number (0–6) used for sorting |
| `weekday_name` | Day name (Monday–Sunday) used for trend analysis |
| `quarter` | Quarter of the year (Q1–Q4) |

### Purpose

These columns were created to:

- Analyze revenue trends by month and year.
- Compare sales performance across weekdays.
- Enable proper sorting of months and weekdays in visualizations.
- Support quarterly business performance analysis.
- Improve dashboard interactivity through date-based filtering and slicing.


```python
df['year'] = df['order_date'].dt.year
```

```python
df['month_num'] = df['order_date'].dt.month
```

```python
df['month_name'] = df['order_date'].dt.month_name()
```

```python
df['weekday_num'] = df['order_date'].dt.weekday
```

```python
df['weekday_name'] = df['order_date'].dt.day_name()
```

```python
df['quarter'] = df['order_date'].dt.quarter
```

```python
#Monthly-Revenue-Trend
monthly_revenue = (
    df.groupby(['year','month_num','month_name'])['revenue']
      .sum()
      .reset_index()
)

monthly_revenue
```

```python
#Weekday-Revenue

weekday_revenue = (
    df.groupby(['weekday_num','weekday_name'])['revenue']
      .sum()
      .reset_index()
      .sort_values('weekday_num')
)

weekday_revenue
```

```python
#Revenue-By-Country

country_revenue = (
    df.groupby('country')['revenue']
      .sum()
      .sort_values(ascending=False)
)

country_revenue
```

```python
#Revenue-By-Channel


channel_revenue = df.groupby('channel')['revenue'].sum().sort_values(ascending = False)

channel_revenue
```

```python
#Boxes-shipped-by-channel

channel_boxes = (df.groupby('channel')['boxes_shipped']
               .sum()
               .sort_values(ascending=False))
channel_boxes
```

```python
#Top-products-by-revenue

product_revenue = (
    df.groupby('product')['revenue']
      .sum()
      .sort_values(ascending=False)
)

product_revenue.head(10)
```

```python
#Top-sales-persons

salesperson_revenue = (
    df.groupby('salesperson')['revenue']
      .sum()
      .sort_values(ascending=False)
)

salesperson_revenue.head(10)
```

- Discounts show only a weak relationship with shipment volume (**0.17**).
